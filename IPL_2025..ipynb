IPL 2025 Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
dfu=pd.read_csv("matches.csv")
dfu
```

```python
dfu.isnull().sum()
```

```python
dfu.dropna(how="all")
```

Since columns- wb_runs and wb_wickets contains too many null values, we are droping those columns from dataset.

Missing values are absolute values, filling them up will reduce the data quality and meaning of the datset hence continuing with analysis.

```python
dfu.drop(['wb_runs','wb_wickets'],axis=1,inplace=True)
```

```python
dfu.head(5)
```

Which team won the most mathches?

```python
winners=dfu['match_winner'].value_counts()
sns.barplot(y=winners.index,x=winners.values,palette='viridis')
plt.title('Most matches won by teams')
```

toss decision trend

```python
toss_trend=dfu['toss_decision'].value_counts()
toss_trend
```

```python
sns.barplot(x=toss_trend.index,y=toss_trend.values,palette='viridis')
```

```python
sns.countplot(dfu['toss_decision'],color='hotpink')
plt.title("toss_trend")
```

Toss Winners vs Match Winners

```python
count=dfu[dfu['toss_winner']==dfu['match_winner']]['match_id'].count() #Teams that won toss and won match
```

```python
dfu[dfu['toss_winner']!=dfu['match_winner']]['match_id'].count() #Teams that wont toss and DID NOT won match
```

```python
percentage=(count*100)/dfu.shape[0]
percentage.round(2)
```

Most Man of the match award winner

```python
ManOfMatch=dfu['player_of_the_match'].value_counts().head(10)
ManOfMatch
```

```python
manAward=sns.barplot(x=ManOfMatch.values,y=ManOfMatch.index,palette='viridis')
plt.title('Top 10 Man of the Match Award holders')
```

Top Scorers

```python
top_s=dfu.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_s.values,y=top_s.index,palette='rainbow')
plt.title("Top 10 Scorer of the IPL 2025")
```

10 Best Bowling Figures

```python
dfu['highest_wickets'] = dfu['best_bowling_figure'].apply(lambda x: str(x).split('--')[0] if pd.notnull(x) else '0')
dfu['highest_wickets'] = dfu['highest_wickets'].astype(int)
```

```python
top_bowlingFig=dfu.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlingFig

top_bowlingFig.plot(kind='barh',colormap='rainbow')
plt.title('Top 10 best bowlig Figures in terms of Wickets taken')
```

Most matches played on a Venue.

```python
venue=dfu['venue'].value_counts().head(5)
venue
venue.plot(kind='barh',x=venue.values,y=venue.index,colormap='rainbow')
plt.title('Most matches played on a venue')
```
